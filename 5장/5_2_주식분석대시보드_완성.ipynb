````python
%%writefile stock_analyzer.py
import streamlit as st
import yfinance as yf
import pandas as pd
import plotly.graph_objects as go
import numpy as np

st.set_page_config(
    page_title="ì£¼ì‹ ë¶„ì„ê¸°",
    page_icon="ğŸ“ˆ", 
    layout="wide"
)

STOCKS = {
    'AAPL': 'Apple Inc.',
    'MSFT': 'Microsoft Corp.',
    'GOOGL': 'Alphabet Inc.',
    'TSLA': 'Tesla Inc.',
    'NVDA': 'NVIDIA Corp.'
}

def get_stock_data(symbol, period="1y"):
    """
    ì£¼ì‹ ë°ì´í„°ë¥¼ ê°€ì ¸ì˜¤ëŠ” í•¨ìˆ˜

    Parameters:
    symbol (str): ì£¼ì‹ ì‹¬ë³¼ (ì˜ˆ: 'AAPL')
    period (str): ë°ì´í„° ê¸°ê°„ (ì˜ˆ: '1y', '6mo', '3mo')

    Returns:
    tuple: (ì—­ì‚¬ì  ë°ì´í„°, ê¸°ì—… ì •ë³´) ë˜ëŠ” (None, None)
    """
    try:
        # yfinance Ticker ê°ì²´ ìƒì„±
        ticker = yf.Ticker(symbol)  # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš”

        # ì—­ì‚¬ì  ë°ì´í„° ê°€ì ¸ì˜¤ê¸°
        hist_data = ticker.history(period=period)  # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš”

        # ê¸°ì—… ì •ë³´ ê°€ì ¸ì˜¤ê¸°
        company_info = ticker.info  # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš”

        # ë°ì´í„°ê°€ ë¹„ì–´ìˆì§€ ì•Šìœ¼ë©´ ë°˜í™˜
        if not hist_data.empty:
            return hist_data, company_info
        else:
            return None, None

    except Exception as e:
        st.error(f"ë°ì´í„°ë¥¼ ê°€ì ¸ì˜¤ëŠ” ì¤‘ ì˜¤ë¥˜ ë°œìƒ: {e}")
        return None, None

def display_company_info(company_info, current_price):
    """
    ê¸°ì—… ê¸°ë³¸ ì •ë³´ë¥¼ í‘œì‹œí•˜ëŠ” í•¨ìˆ˜

    Parameters:
    company_info (dict): ê¸°ì—… ì •ë³´ ë”•ì…”ë„ˆë¦¬
    current_price (float): í˜„ì¬ ì£¼ê°€
    """
    # 4ê°œì˜ ì—´ë¡œ ë‚˜ëˆ„ê¸°
    col1, col2, col3, col4 = st.columns(4)  # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš”

    with col1:
        st.metric("í˜„ì¬ê°€", f"${current_price:.2f}")

    with col2:
        # ì‹œê°€ì´ì•¡ ê°€ì ¸ì˜¤ê¸° (ì—†ìœ¼ë©´ 0)
        market_cap = company_info.get('marketCap', 0)  # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš”
        st.metric("ì‹œê°€ì´ì•¡", f"${market_cap:,.0f}")

    with col3:
        # PER ê°€ì ¸ì˜¤ê¸° (ì—†ìœ¼ë©´ 0)
        pe_ratio = company_info.get('trailingPE', 0)  # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš” (hint: trailing + PE)
        st.metric("PER", f"{pe_ratio:.2f}" if pe_ratio else "N/A")

    with col4:
        # ë°°ë‹¹ìˆ˜ìµë¥  ê°€ì ¸ì˜¤ê¸°
        dividend_yield = company_info.get('dividendYield', 0)
        if dividend_yield:
            st.metric("ë°°ë‹¹ìˆ˜ìµë¥ ", f"{dividend_yield*100:.2f}%")
        else:
            st.metric("ë°°ë‹¹ìˆ˜ìµë¥ ", "N/A")

def create_price_chart(hist_data, symbol):
    """
    ì£¼ê°€ ìº”ë“¤ìŠ¤í‹± ì°¨íŠ¸ë¥¼ ìƒì„±í•˜ëŠ” í•¨ìˆ˜

    Parameters:
    hist_data (DataFrame): ì£¼ì‹ ì—­ì‚¬ì  ë°ì´í„°
    symbol (str): ì£¼ì‹ ì‹¬ë³¼

    Returns:
    plotly.graph_objects.Figure: ì°¨íŠ¸ ê°ì²´
    """
    # ë¹ˆ Figure ê°ì²´ ìƒì„±
    fig = go.Figure()  # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš”

    # ìº”ë“¤ìŠ¤í‹± ì°¨íŠ¸ ì¶”ê°€
    fig.add_trace(go.Candlestick(
        x=hist_data.index,
        open=hist_data['Open'],    # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš” (ì‹œê°€)
        high=hist_data['High'],    # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš” (ê³ ê°€)
        low=hist_data['Low'],     # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš” (ì €ê°€)
        close=hist_data['Close'],   # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš” (ì¢…ê°€)
        name=symbol
    ))

    # 20ì¼ ì´ë™í‰ê· ì„  ì¶”ê°€ (ë°ì´í„°ê°€ 20ì¼ ì´ìƒì¼ ë•Œ)
    if len(hist_data) >= 20:
        ma20 = hist_data['Close'].rolling(window=20).mean()  # ë¹ˆì¹¸ 2ê°œë¥¼ ì±„ìš°ì„¸ìš”
        fig.add_trace(go.Scatter(
            x=hist_data.index,
            y=ma20,
            mode='lines',
            name='MA20',
            line=dict(color='orange', width=1)
        ))

    # ì°¨íŠ¸ ë ˆì´ì•„ì›ƒ ì„¤ì •
    fig.update_layout(  # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš”
        title=f"{symbol} ì£¼ê°€ ì°¨íŠ¸",
        yaxis_title="ê°€ê²© ($)",
        xaxis_title="ë‚ ì§œ",
        height=500
    )

    return fig


def create_volume_chart(hist_data, symbol):
    """
    ê±°ë˜ëŸ‰ ì°¨íŠ¸ë¥¼ ìƒì„±í•˜ëŠ” í•¨ìˆ˜

    Parameters:
    hist_data (DataFrame): ì£¼ì‹ ì—­ì‚¬ì  ë°ì´í„°
    symbol (str): ì£¼ì‹ ì‹¬ë³¼

    Returns:
    plotly.graph_objects.Figure: ì°¨íŠ¸ ê°ì²´
    """
    fig = go.Figure()

    # ê±°ë˜ëŸ‰ ë§‰ëŒ€ ì°¨íŠ¸ ì¶”ê°€
    fig.add_trace(go.Bar(  # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš” (Bar ì°¨íŠ¸)
        x=hist_data.index,
        y=hist_data['Volume'],  # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš” (ê±°ë˜ëŸ‰ ì»¬ëŸ¼)
        name='ê±°ë˜ëŸ‰',
        marker_color='lightblue'
    ))

    # ì°¨íŠ¸ ë ˆì´ì•„ì›ƒ ì„¤ì •
    fig.update_layout(
        title=f"{symbol} ê±°ë˜ëŸ‰",
        xaxis_title="ë‚ ì§œ",
        yaxis_title="ê±°ë˜ëŸ‰",
        height=300
    )

    return fig


def calculate_technical_indicators(hist_data):
    """
    ê¸°ìˆ ì  ì§€í‘œë¥¼ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜

    Parameters:
    hist_data (DataFrame): ì£¼ì‹ ì—­ì‚¬ì  ë°ì´í„°

    Returns:
    dict: ê³„ì‚°ëœ ì§€í‘œë“¤ì˜ ë”•ì…”ë„ˆë¦¬
    """
    # ì¼ì¼ ìˆ˜ìµë¥  ê³„ì‚° (ì „ì¼ ëŒ€ë¹„ ë³€í™”ìœ¨)
    returns = hist_data['Close'].pct_change().dropna()  # ë¹ˆì¹¸ 2ê°œë¥¼ ì±„ìš°ì„¸ìš”

    # ê¸°ê°„ ìˆ˜ìµë¥  ê³„ì‚° (ì „ì²´ ê¸°ê°„)
    total_return = ((hist_data['Close'].iloc[-1] / hist_data['Close'].iloc[0]) - 1) * 100  # ë¹ˆì¹¸ 2ê°œ

    # ì¼ì¼ í‰ê·  ìˆ˜ìµë¥ 
    avg_daily_return = returns.mean() * 100  # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš”

    # ì—°ê°„ ë³€ë™ì„± (ì¼ì¼ ë³€ë™ì„± Ã— âˆš252)
    volatility = returns.std() * np.sqrt(252) * 100  # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš”

    # ìµœëŒ€ ì†ì‹¤í­ (MDD) ê³„ì‚°
    cummax = hist_data['Close'].cummax()  # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš” (ëˆ„ì  ìµœëŒ€ê°’)
    drawdown = (cummax - hist_data['Close']) / cummax
    max_drawdown = drawdown.max() * 100

    # ìƒ¤í”„ ë¹„ìœ¨ (ìœ„í—˜ ëŒ€ë¹„ ìˆ˜ìµë¥ )
    sharpe_ratio = (avg_daily_return / returns.std() * np.sqrt(252)) if returns.std() > 0 else 0

    return {
        'total_return': total_return,
        'avg_daily_return': avg_daily_return,
        'volatility': volatility,
        'max_drawdown': max_drawdown,
        'sharpe_ratio': sharpe_ratio
    }


def main():
    st.title("ğŸ“ˆ ì£¼ì‹ ë¶„ì„ ëŒ€ì‹œë³´ë“œ")

    selected_symbol = st.selectbox(
        "ë¶„ì„í•  ì¢…ëª©ì„ ì„ íƒí•˜ì„¸ìš”:",
        list(STOCKS.keys()),
        format_func=lambda x: f"{x} - {STOCKS[x]}"
    )

    period_options = {
        "3ê°œì›”": "3mo",
        "6ê°œì›”": "6mo", 
        "1ë…„": "1y",
        "2ë…„": "2y"
    }

    selected_period = st.selectbox("ë¶„ì„ ê¸°ê°„:", list(period_options.keys()))
    period_code = period_options[selected_period]

    with st.spinner("ë°ì´í„°ë¥¼ ê°€ì ¸ì˜¤ëŠ” ì¤‘..."):
        hist_data, company_info = get_stock_data(selected_symbol, period_code)

    if hist_data is None:
        st.error("ë°ì´í„°ë¥¼ ê°€ì ¸ì˜¬ ìˆ˜ ì—†ìŠµë‹ˆë‹¤.")
        return

    current_price = hist_data['Close'].iloc[-1]

    display_company_info(company_info, current_price)

    st.subheader("ğŸ“ˆ ì£¼ê°€ ì°¨íŠ¸")
    price_chart = create_price_chart(hist_data, selected_symbol)
    st.plotly_chart(price_chart, use_container_width=True)

    st.subheader("ğŸ“Š ê±°ë˜ëŸ‰")
    volume_chart = create_volume_chart(hist_data, selected_symbol)
    st.plotly_chart(volume_chart, use_container_width=True)

    st.subheader("ğŸ“Š ê¸°ìˆ ì  ì§€í‘œ")
    indicators = calculate_technical_indicators(hist_data)

    metrics_df = pd.DataFrame({
        "ì§€í‘œ": ["ì´ ìˆ˜ìµë¥ ", "ì¼ì¼ í‰ê·  ìˆ˜ìµë¥ ", "ì—°ê°„ ë³€ë™ì„±", "ìµœëŒ€ ì†ì‹¤í­", "ìƒ¤í”„ ë¹„ìœ¨"],
        "ê°’": [
            f"{indicators['total_return']:.2f}%",
            f"{indicators['avg_daily_return']:.3f}%", 
            f"{indicators['volatility']:.2f}%",
            f"{indicators['max_drawdown']:.2f}%",
            f"{indicators['sharpe_ratio']:.2f}"
        ]
    })

    st.dataframe(metrics_df, use_container_width=True)

if __name__ == "__main__":
    main()
````

````python
def calculate_technical_indicators(hist_data):
    """
    ê¸°ìˆ ì  ì§€í‘œë¥¼ ê³„ì‚°í•˜ëŠ” í•¨ìˆ˜

    Parameters:
    hist_data (DataFrame): ì£¼ì‹ ì—­ì‚¬ì  ë°ì´í„°

    Returns:
    dict: ê³„ì‚°ëœ ì§€í‘œë“¤ì˜ ë”•ì…”ë„ˆë¦¬
    """
    # ì¼ì¼ ìˆ˜ìµë¥  ê³„ì‚° (ì „ì¼ ëŒ€ë¹„ ë³€í™”ìœ¨)
    returns = hist_data['Close'].pct_change().dropna()  # ë¹ˆì¹¸ 2ê°œë¥¼ ì±„ìš°ì„¸ìš”

    # ê¸°ê°„ ìˆ˜ìµë¥  ê³„ì‚° (ì „ì²´ ê¸°ê°„)
    total_return = ((hist_data['Close'].iloc[-1] / hist_data['Close'].iloc[0]) - 1) * 100  # ë¹ˆì¹¸ 2ê°œ

    # ì¼ì¼ í‰ê·  ìˆ˜ìµë¥ 
    avg_daily_return = returns.mean() * 100  # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš”

    # ì—°ê°„ ë³€ë™ì„± (ì¼ì¼ ë³€ë™ì„± Ã— âˆš252)
    volatility = returns.std() * np.sqrt(252) * 100  # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš”

    # ìµœëŒ€ ì†ì‹¤í­ (MDD) ê³„ì‚°
    cummax = hist_data['Close'].cummax()  # ë¹ˆì¹¸ì„ ì±„ìš°ì„¸ìš” (ëˆ„ì  ìµœëŒ€ê°’)
    drawdown = (cummax - hist_data['Close']) / cummax
    max_drawdown = drawdown.max() * 100

    # ìƒ¤í”„ ë¹„ìœ¨ (ìœ„í—˜ ëŒ€ë¹„ ìˆ˜ìµë¥ )
    sharpe_ratio = (avg_daily_return / returns.std() * np.sqrt(252)) if returns.std() > 0 else 0

    return {
        'total_return': total_return,
        'avg_daily_return': avg_daily_return,
        'volatility': volatility,
        'max_drawdown': max_drawdown,
        'sharpe_ratio': sharpe_ratio
    }
````
