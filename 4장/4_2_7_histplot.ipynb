# 히스토그램(histplot)

```python
import seaborn as sns
import pandas as pd
```

```python
df = sns.load_dataset("tips")
df.head()
```

## 1. 기본 히스토그램 생성

```python
sns.histplot(data=df['total_bill'])
```

## 2. bins(구간)조정

```python
sns.histplot(data=df['total_bill'], bins=5)
```

## 3. KDE 활용

```python
sns.histplot(data=df['total_bill'], kde=True)
```

# 분포 시각화를 위한 다른 방법

## 1. KDE plot

```python
sns.kdeplot(data=df['total_bill'])
```

## 2. rug plot

```python
sns.rugplot(data=df['total_bill'])
```

## 3. 히스토그램 vs 카운트플롯

```python
sns.histplot(data=df['day'])
```

```python
sns.countplot(data=df['day'])
```
