# 회귀선(regplot)

```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset("tips")
```

## 기본 회귀선 그래프

```python
sns.regplot(x=df['total_bill'], y=df['tip'])
```

## 회귀선만 표시하기

```python
sns.regplot(x=df['total_bill'], y=df['tip'], scatter = False)
```

## 신뢰구간의 이해

```python
sns.regplot(x=df['total_bill'], y=df['tip'], scatter = False, ci = None)
```
